8.Hur 收盘价减去上一个长度为20天的区间的平均收盘价在过去二十天的和（参考广发金工20170921研报）

```python
def interval_mean(Df,n):
    import numpy as np
    Df_mean=Df.rolling(n).mean().iloc[::n]
    Df_copy=Df.copy()
    Df_copy.loc[:,:]=np.nan
    Df_copy.loc[Df_mean.index]=Df_mean
    Df_copy=Df_copy.fillna(method='ffill')
    #注意这里要用ffill不能用bfill，用bfill就用到未来值了
    return Df_copy
```

```python
dv.add_formula('Hur','Ts_Sum(close-interval_mean(close,20),20)',is_quarterly=False,add_data=True,register_funcs={'interval_mean':interval_mean}).head()
```

11.crosssec_weight_vc 成交量（volume)权重为2的20天SMA的横截面权重与收盘价(close)权重为2的20天SMA的横截面权重的乘积

```python
def sum_axis1(Df):
    sum_1=Df.sum(axis=1)
    Df_copy=Df.copy()
    Df_copy.loc[sum_1.index,:]=sum_1.values.reshape(-1,1)
    return Df_copy
```

```python
def SMA(Df,n,m):
    alpha=m/n
    return Df.ewm(alpha=alpha,adjust=False).mean()
```

```python
dv.add_formula('crosssec_weight_vc','SMA(volume,20,2)/sum_axis1(SMA(volume,20,2))*SMA(close,20,2)/sum_axis1(SMA(close,20,2))',is_quarterly=False,add_data=True,register_funcs={'SMA':SMA,'sum_axis1':sum_axis1}).tail()
```

12.Factor1 以收盘价横截面比值为权重的最高价和vwap的加权和在过去四天的变化值的横截面排序

```python
def sum_axis1(Df):
    sum_1=Df.sum(axis=1)
    Df_copy=Df.copy()
    Df_copy.loc[sum_1.index,:]=sum_1.values.reshape(-1,1)
    return Df_copy
```

```python
dv.add_formula('Factor1','Rank(Delta(((high*close/sum_axis1(close)) + (vwap *(1-close/sum_axis1(close)))), 4) )',is_quarterly=False,add_data=True,register_funcs={'sum_axis1':sum_axis1}).tail()
```

```python
import numpy as np

def limit_up_down():
    trade_status = dv.get_ts('trade_status').fillna(0)
    mask_sus = trade_status == 0
    up_limit = dv.add_formula('up_limit', '(close - Delay(close, 1)) / Delay(close, 1) > 0.095', is_quarterly=False)
    down_limit = dv.add_formula('down_limit', '(close - Delay(close, 1)) / Delay(close, 1) < -0.095', is_quarterly=False)
    can_enter = np.logical_and(up_limit < 1, ~mask_sus) 
    can_exit = np.logical_and(down_limit < 1, ~mask_sus) 
    return can_enter,can_exit
```

```python
can_enter,can_exit=limit_up_down()
```

```python
price=dv.get_ts('close_adj')
sw1=dv.get_ts('sw1')
enter=can_enter
exit=can_exit
mask=mask
```

```python
import matplotlib.pyplot as plt
from jaqs_fxdayu.research import SignalDigger
from jaqs_fxdayu.research.signaldigger import analysis

def cal_obj(signal, name, period, quantile):
    obj = SignalDigger(output_folder="ZZ800/%s" % name,
                       output_format='pdf')
    obj.process_signal_before_analysis(signal,
                                   price=price,
                                   n_quantiles=quantile, period=period,
                                   mask=mask,
                                   group=sw1,
                                   can_enter = enter,
                                   can_exit = exit,
                                   commission = 0.0008
                                   )
    #obj.create_full_report()
    return obj

def plot_pfm(signal, name, period=5, quantile=5):
    obj = cal_obj(signal, name, period, quantile)
    plt.show()
def signal_data(signal, name, period=5, quantile=5):
    obj = cal_obj(signal, name, period, quantile)
    return obj.signal_data
```

```python
from jaqs_fxdayu.research.signaldigger.process import neutralize
```

```python
def ic_length(signal, days=750):
    return signal.loc[signal.index.levels[0][-days]:]
```

```python
from jaqs_fxdayu.research.signaldigger import performance as pfm
```

测试因子的IC和IR

```python
alpha_signal=['PBHist120','turnover_mean20_sd','high_length_minus_low_length','volume_ratio',
              'ETOPHist120','TD_intermediate','psHist120',
              'Hur','close_ratio','LCAPHist120','crosssec_weight_vc','Factor1','Factor2']
```

```python
neutralize_dict={a:neutralize(factor_df=dv.get_ts(a),group=dv.get_ts('sw1')) for a in alpha_signal}
signals_dict = {a:signal_data(neutralize_dict[a], a, 20) for a in alpha_signal} 
ic_pn = pd.Panel({a: analysis.ic_stats(signals_dict[a]) for a in signals_dict.keys()})
alpha_performance=abs(round(ic_pn.minor_xs('return_ic'),2))
alpha_IR = alpha_performance.loc["Ann. IR"]
alpha_IC = alpha_performance.loc["IC Mean"]
good_alpha = alpha_IC[(alpha_IC>=0.03) & (alpha_IR>=0.25)]
good_alpha_dict = {g: float('%.2f' % good_alpha[g]) for g in good_alpha.index}
good_alpha_dict
```

找出高IC行业

```python
signal_dict_industry = {alpha : signal_data(dv.get_ts(alpha), alpha, period=20, quantile=5) for alpha in good_alpha.index}
performance_dict = {}
for alpha in good_alpha.index:
    ic = pfm.calc_signal_ic(ic_length(signal_dict_industry[alpha]), by_group=True)
    mean_ic_by_group = pfm.mean_information_coefficient(ic, by_group=True)
    performance_dict[alpha] = round(mean_ic_by_group,2)
ic_industry=abs(pd.Panel(performance_dict).minor_xs('ic'))
High_IC_Industry = pd.DataFrame([ic_industry[ic_industry>=0.05][alpha].dropna(how='all') for alpha in good_alpha.index]).T
```
